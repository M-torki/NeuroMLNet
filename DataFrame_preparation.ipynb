```python
control_adults = sorted(glob('./energies/*control_adulthood.txt'))

df_control_adults = pd.read_csv(control_adults[0], names=[control_adults[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(control_adults)):

    df = pd.read_csv(control_adults[i], names=[control_adults[i].split('/')[-1].split('_')[-3]])
    df_control_adults = pd.concat([df_control_adults , df] , axis=1)
    # df_control_adults.merge(df , how='left' , left_index=True)

df_control_adults.head()
```

```python
df_control_adults.shape
```

```python
df_control_adults['Age'] = 2
df_control_adults['Label'] = 0
print(df_control_adults.shape)
df_control_adults.head()
```

```python
control_adols = sorted(glob('./energies/*control_adolescence.txt'))

df_control_adols = pd.read_csv(control_adols[0], names=[control_adols[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(control_adols)):

    df = pd.read_csv(control_adols[i], names=[control_adols[i].split('/')[-1].split('_')[-3]])
    df_control_adols = pd.concat([df_control_adols , df] , axis=1)

df_control_adols['Age'] = 1
df_control_adols['Label'] = 0
print(df_control_adols.shape)
# df_control_adols.head()
```

```python
control_childs = sorted(glob('./energies/*control_childhood.txt'))

df_control_childs = pd.read_csv(control_childs[0], names=[control_childs[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(control_childs)):

    df = pd.read_csv(control_childs[i], names=[control_childs[i].split('/')[-1].split('_')[-3]])
    df_control_childs = pd.concat([df_control_childs , df] , axis=1)

df_control_childs['Age'] = 0
df_control_childs['Label'] = 0
print(df_control_childs.shape)
# df_control_childs.head()
```

```python
autism_adults = sorted(glob('./energies/*autism_adulthood.txt'))

df_autism_adults = pd.read_csv(autism_adults[0], names=[autism_adults[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(autism_adults)):

    df = pd.read_csv(autism_adults[i], names=[autism_adults[i].split('/')[-1].split('_')[-3]])
    df_autism_adults = pd.concat([df_autism_adults , df] , axis=1)
    # df_control_adults.merge(df , how='left' , left_index=True)

df_autism_adults['Age'] = 2
df_autism_adults['Label'] = 1
print(df_autism_adults.shape)
# df_autism_adults.head()
```

```python
autism_adols = sorted(glob('./energies/*autism_adolescence.txt'))

df_autism_adols = pd.read_csv(autism_adols[0], names=[autism_adols[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(autism_adols)):

    df = pd.read_csv(autism_adols[i], names=[autism_adols[i].split('/')[-1].split('_')[-3]])
    df_autism_adols = pd.concat([df_autism_adols , df] , axis=1)

df_autism_adols['Age'] = 1
df_autism_adols['Label'] = 1
print(df_autism_adols.shape)
# df_autism_adols.head()
```

```python
autism_childs = sorted(glob('./energies/*autism_childhood.txt'))

df_autism_childs = pd.read_csv(autism_childs[0], names=[autism_childs[0].split('/')[-1].split('_')[-3]])

for i in range(1,len(autism_childs)):

    df = pd.read_csv(autism_childs[i], names=[autism_childs[i].split('/')[-1].split('_')[-3]])
    df_autism_childs = pd.concat([df_autism_childs , df] , axis=1)

df_autism_childs['Age'] = 0
df_autism_childs['Label'] = 1
print(df_autism_childs.shape)
# df_autism_childs.head()
```

```python
df_tot = pd.concat([df_control_adults , df_control_adols , df_control_childs , df_autism_adults , df_autism_adols , df_autism_childs] , axis=0)
print(df_tot.shape)
# df_tot.head()
```

```python
df_tot = df_tot.sample(frac=1).reset_index(drop=True)
df_tot.head()
```

```python
df_tot.to_csv('df_tot.csv' , index=False)
```
